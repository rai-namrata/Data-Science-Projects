```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
len_of_lists=[len(stem_cats),len(lib_arts_cats),len(other_cats)]
list_of_lists=[stem_cats,lib_arts_cats,other_cats]

fig = plt.figure(figsize=(18, 18))

for i in range(len(len_of_lists)):
    j=0
    x=i*6
    sp=i+1
    for itr in range(x,x+len_of_lists[i]):     
        ax = fig.add_subplot(6,3,sp)
        ax.plot(women_degrees['Year'], women_degrees[list_of_lists[i][j]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[list_of_lists[i][j]], c=cb_orange, label='Men', linewidth=3)
        ax.spines["right"].set_visible(False)    
        ax.spines["left"].set_visible(False)
        ax.spines["top"].set_visible(False)    
        ax.spines["bottom"].set_visible(False)
        ax.set_xlim(1968, 2011)
        # ax.set_ylim(0,100) #uncommenting won't make a difference to the plots
        ax.set_yticks([0,100])
        ax.set_title(list_of_lists[i][j])
        ax.tick_params(bottom=False, top=False, left=False, right=False,labelbottom=False)
        ax.axhline(50,c=(171/255,183/255,171/255),alpha=0.3)
        if i==0:
            if sp==1:
                ax.text(2005,81,'Women')
                ax.text(2007,15,'Men')
            elif sp==16:
                ax.text(2007,88,'Men')
                ax.text(2005,9,'Women')
                ax.tick_params(labelbottom=True)
        elif i==1:
            if sp==2:
                ax.text(2005,76,'Women')
                ax.text(2007,22,'Men')
            if sp==14:
                ax.tick_params(labelbottom=True)
        elif i==2:
            if sp==3:
                ax.text(2005,90,'Women')
                ax.text(2007,8,'Men')
            elif sp==18:
                ax.text(2007,62,'Men')
                ax.text(2005,36,'Women')
                ax.tick_params(labelbottom=True)
    
        j+=1
        sp=sp+3
        
plt.savefig("gender_degrees.png")
#plt.tight_layout() # another way to format the plots and improve there appearence
plt.show()
```
